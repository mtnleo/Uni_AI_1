**Ejercicio 1: Utilizando la función make_blobs de sklearn.datasets import make_blobs
1.  Genere un conjunto de datos sinteticos de 400 muestras pertenecientes a 4 grupos o clases.
2.  Guardelos en un dataframe con sus respectivas columnas identificadas.
3. Visualize los datos sin las etiquetas
4. Explique que es estandarizar y normalizar- escriba las ecuaciones relacionadas.
5. Estandarizar los datos con media 0 y desviacion estandar 1

---

```python
from sklearn.datasets import make_blobs # 1.
from sklearn.preprocessing import StandardScaler, MinMaxScaler # 5.

# 1. Genere un conjunto de datos sinteticos de 400 muestras pertenecientes a 4
# grupos o clases.
X_blobs, y_blobs = make_blobs(n_samples=400, centers=4,
                               cluster_std=0.60, random_state=0)
# 2. Guardelos en un dataframe con sus respectivas columnas identificadas.
df_blobs = pd.DataFrame(X_blobs, columns=['Feature1', 'Feature2'])
df_blobs['Cluster_Real'] = y_blobs

# 3. Visualice los datos sin las etiquetas
print("Datos para clustering:")
df_blobs.head()


plt.figure(figsize=(12, 5)) # Ver datos

plt.subplot(1, 2, 1)
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], s=50)
plt.title('Datos sin Etiquetar')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()

# 4. Explique que es estandarizar y normalizar- escriba las ecuaciones relacionadas.
"""
Estandarización (Z-score):
Transforma los datos para que tengan una media de 0 y una desviación estándar
de 1. Es ideal para algoritmos que asumen una distribución normal o gaussiana
(como SVM o Regresión Lineal).

Ecuación: z = (x - media) / desviación

Normalización (Min-Max):
Reescala los datos para que se encuentren en un rango específico, generalmente
 entre 0 y 1. Es útil para algoritmos que no asumen una distribución específica
y son sensibles a la magnitud, como Redes Neuronales o KNN.

Ecuación: xnorm = (x - xmin) / (xmax - xmin)

"""

# 5. Estandarizar los datos con media 0 y desviacion estandar 1
scaler_std = StandardScaler()
X_blobs_std = scaler_std.fit_transform(X_blobs)

print("Después de estandarización (media=0, std=1):")
print(f"Media: {X_blobs_std.mean(axis=0).round(2)}")
print(f"Desviación: {X_blobs_std.std(axis=0).round(2)}")


```
